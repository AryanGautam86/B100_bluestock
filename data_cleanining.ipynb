```python
import pandas as pd
import numpy as np
import os

raw_path = "data/"
clean_path = "cleaned_data/"

os.makedirs(clean_path, exist_ok=True)

for file in os.listdir(raw_path):

    if not file.endswith(".xlsx"):
        continue

    print(f"Processing {file}")

    df = pd.read_excel(
        os.path.join(raw_path, file),
        skiprows=1,          # Skip title row
        engine="openpyxl"
    )

    # Clean column names
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("-", "_")
    )

    # Convert NULL strings to NaN
    df.replace(
        ["NULL", "Null", "null", ""],
        np.nan,
        inplace=True
    )

    # Remove fully empty rows
    df.dropna(how="all", inplace=True)

    # Remove duplicate rows
    df.drop_duplicates(inplace=True)

    # Save cleaned file
    output_file = file.replace(".xlsx", ".csv")

    df.to_csv(
        os.path.join(clean_path, output_file),
        index=False
    )

    print(f"Saved -> cleaned_data/{output_file}")
    print(f"Shape -> {df.shape}")

print("All files cleaned successfully.")
```

```python
import pandas as pd
import numpy as np
import os

# =========================
# PATHS
# =========================

INPUT_FOLDER = "data"
OUTPUT_FOLDER = "data/cleaned"

os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# =========================
# PROCESS ALL XLSX FILES
# =========================

for file in os.listdir(INPUT_FOLDER):

    if not file.endswith(".xlsx"):
        continue

    print(f"\nProcessing {file}")

    file_path = os.path.join(INPUT_FOLDER, file)

    # Read Excel
    df = pd.read_excel(
        file_path,
        header=1,  # your files have title row first
        engine="openpyxl"
    )

    # =========================
    # CLEAN COLUMN NAMES
    # =========================

    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.lower()
        .str.replace(" ", "_", regex=False)
        .str.replace("-", "_", regex=False)
        .str.replace("/", "_", regex=False)
    )

    # =========================
    # REMOVE UNNAMED COLUMNS
    # =========================

    df = df.loc[
        :,
        ~df.columns.str.contains("^unnamed", case=False)
    ]

    # =========================
    # REPLACE NULL VALUES
    # =========================

    df.replace(
        ["NULL", "Null", "null", ""],
        np.nan,
        inplace=True
    )

    # =========================
    # STRIP STRING COLUMNS
    # =========================

    for col in df.columns:

        if df[col].dtype == "object":

            df[col] = (
                df[col]
                .astype(str)
                .str.strip()
            )

    # =========================
    # COMPANIES FILE SPECIAL FIX
    # =========================

    if file.lower() == "companies.xlsx":

        if "id" in df.columns:

            df.rename(
                columns={"id": "company_id"},
                inplace=True
            )

        if "company_id" in df.columns:

            df["company_id"] = (
                df["company_id"]
                .astype(str)
                .str.strip()
                .str.upper()
            )

    # =========================
    # OTHER FILES
    # =========================

    if "company_id" in df.columns:

        df["company_id"] = (
            df["company_id"]
            .astype(str)
            .str.strip()
            .str.upper()
        )

    # =========================
    # REMOVE EMPTY ROWS
    # =========================

    df.dropna(how="all", inplace=True)

    # =========================
    # REMOVE DUPLICATES
    # =========================

    df.drop_duplicates(inplace=True)

    # =========================
    # SAVE CSV
    # =========================

    output_file = file.replace(".xlsx", ".csv")

    output_path = os.path.join(
        OUTPUT_FOLDER,
        output_file
    )

    df.to_csv(
        output_path,
        index=False
    )

    print(f"Saved -> {output_path}")
    print(f"Rows -> {len(df)}")
    print(f"Columns -> {len(df.columns)}")

print("\nAll files converted successfully.")
```
